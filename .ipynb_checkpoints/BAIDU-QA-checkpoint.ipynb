```python
MAX_LEN = 512
TRAIN_BATCH_SIZE = 8
VALID_BATCH_SIZE = 8
EPOCHS = 5
PATH = "bert_config" #！！！需要改
TRAINING_FILE = "train_deleted.csv" #！！！需要改
VALID_FILE = 'dev_data_baidu.csv' #！！！需要改
TOKENIZER = transformers.BertTokenizer.from_pretrained("hfl/chinese-roberta-wwm-ext")
```

```python
def process_data(text, question, answer, tokenizer, max_len):
    len_ans = len(answer)
    idx0 = None
    idx1 = None
    for index in (i for i, e in enumerate(text) if e == answer[0]):
        if text[index:index+len_ans] == answer:
            idx0 = index
            idx1 = index + len_ans - 1
            break
            
    token_text = tokenizer.encode_plus(text,question)
    input_ids_orig = token_text["input_ids"]

    ans_start = idx0+1
    ans_end = idx1+1

    input_ids = input_ids_orig
    #input_ids = [2] + input_ids_orig + [2]
    token_type_ids = [0] * (len(input_ids_orig))
    #token_type_ids = [0] + [0] * (len(input_ids_orig) + 1)
    mask = [1] * len(token_type_ids)
    #ans_start += 1
    #ans_end +=1
    
    padding_length = max_len - len(input_ids)
    if padding_length > 0:
        input_ids = input_ids + ([1] * padding_length)
        mask = mask + [0] * padding_length
        token_type_ids = token_type_ids + [0] * padding_length
    if padding_length < 0:
        input_ids = input_ids[:512]
        mask = mask[:512]
        token_type_ids = token_type_ids[:512]
        ans_start += padding_length
        ans_end += padding_length
        if ans_start < 0 or ans_start>512:
            ans_start = 0
        if ans_end < 0 or ans_end>512:
            ans_end = 0
        
    return {
        'ids': input_ids,
        'mask': mask,
        'token_type_ids': token_type_ids,
        'ans_start': ans_start,
        'ans_end': ans_end,
        'orig_text': text,
        'orig_ans': answer
    }
```

```python
class QAdataset:
    def __init__(self, text,question, answer):
        self.text = text
        self.question = question
        self.answer = answer
        self.tokenizer = TOKENIZER
        self.max_len = MAX_LEN
        
    def __len__(self):
        return len(self.text)
    
    def __getitem__(self, item):
        data = process_data(
            self.text[item],
            self.question[item],
            self.answer[item],
            self.tokenizer,
            self.max_len
            )
        
        return{
            'ids': torch.tensor(data["ids"], dtype=torch.long),
            'mask': torch.tensor(data["mask"], dtype=torch.long),
            'token_type_ids': torch.tensor(data["token_type_ids"], dtype=torch.long),
            'ans_start': torch.tensor(data["ans_start"], dtype=torch.long),
            'ans_end': torch.tensor(data["ans_end"], dtype=torch.long),
            'orig_text': data["orig_text"],
            'orig_ans': data["orig_ans"]
        }
```

```python
class QAmodel(transformers.BertModel):
    def __init__(self, conf):
        super(QAmodel,self).__init__(conf)
        self.bertwwm = transformers.BertModel.from_pretrained("hfl/chinese-roberta-wwm-ext", config=conf)
        self.drop_out = nn.Dropout(0.1)
        self.out = nn.Linear(768 * 2, 2)
        torch.nn.init.normal_(self.out.weight, std=0.02)
        
    def forward(self, input_ids, mask, token_type_ids):
        _, _, out = self.bertwwm(input_ids, token_type_ids=token_type_ids, attention_mask=mask)
        out = torch.cat((out[-1], out[-2]), dim=-1)
        out = self.drop_out(out)
        logits = self.out(out)
        start_logits, end_logits = logits.split(1, dim=-1)
        start_logits = start_logits.squeeze(-1)
        end_logits = end_logits.squeeze(-1)
        return start_logits, end_logits
```

```python
def loss_fn(start_logits, end_logits, start_positions, end_positions):
    loss_fct = nn.CrossEntropyLoss()
    start_loss = loss_fct(start_logits, start_positions)
    end_loss = loss_fct(end_logits, end_positions)
    total_loss = (start_loss + end_loss)
    return total_loss
```

```python
def train_fn(data_loader, model, optimizer, device, scheduler=None):
    model.train()
    for bi, d in enumerate(data_loader):
        ids = d["ids"]
        token_type_ids = d["token_type_ids"]
        mask = d["mask"]
        ans_start = d["ans_start"]
        ans_end = d["ans_end"]
        orig_ans = d["orig_ans"]
        orig_text = d["orig_text"]
        
        ids = ids.to(device, dtype=torch.long)
        token_type_ids = token_type_ids.to(device, dtype=torch.long)
        mask = mask.to(device, dtype=torch.long)
        ans_start = ans_start.to(device, dtype=torch.long)
        ans_end = ans_end.to(device, dtype=torch.long)
        
        model.zero_grad()
        print("ok")
        outputs_start, outputs_end = model(ids, mask, token_type_ids)
        loss = loss_fn(outputs_start, outputs_end, ans_start, ans_end)
        print(loss)
        loss.backward()
        optimizer.step()
        scheduler.step()
        
        outputs_start = torch.softmax(outputs_start, dim=1).cpu().detach().numpy()
        outputs_end = torch.softmax(outputs_end, dim=1).cpu().detach().numpy()
```

```python
df_train = pd.read_csv(TRAINING_FILE)
df_valid = pd.read_csv(VALID_FILE) 
    
train_dataset = QAdataset(
    text=df_train.context.values,
    question=df_train.question.values,
    answer=df_train.answer.values
)

train_data_loader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size=TRAIN_BATCH_SIZE,
)

"""valid_dataset = QAdataset(
    text=df_valid.text.values,
    question=df_vaild.question.values,
    answer=df_valid.answer.values
)

valid_data_loader = torch.utils.data.DataLoader(
    valid_dataset,
    batch_size=1,
)"""

device = torch.device("cuda:1") #!!!需要改成"cuda"

  
model_config = transformers.BertConfig.from_pretrained(PATH)

model_config.output_hidden_states = True

model = QAmodel(conf=model_config)

model.to(device)


num_train_steps = int(len(df_train) / TRAIN_BATCH_SIZE * EPOCHS)

param_optimizer = list(model.named_parameters())

no_decay = ["bias", "LayerNorm.bias", "LayerNorm.weight"]

optimizer_parameters = [
    {'params': [p for n, p in param_optimizer if not any(nd in n for nd in no_decay)], 'weight_decay': 0.001},#if not any(nd in n for nd in no_decay)
    {'params': [p for n, p in param_optimizer if any(nd in n for nd in no_decay)], 'weight_decay': 0.0}#if any(nd in n for nd in no_decay)
]
lr = 3e-5
for epoch in range(3):
    optimizer = AdamW(optimizer_parameters, lr)

    scheduler = get_linear_schedule_with_warmup(
        optimizer, 
        num_warmup_steps=0, 
        num_training_steps=num_train_steps
    )

    #es = utils.EarlyStopping(patience=2, mode="max")
#print(f"Training is Starting for fold")
    
    train_fn(train_data_loader, model, optimizer, device, scheduler=scheduler)
    print("over epoch-------------------------", epoch)
    lr = lr*0.5
torch.save(model, 'QAdev.pt')   
torch.save(model.state_dict(), 'QAdev_params.pt')
```

```python
def eval_fn(data_loader, model, device):
    """
    Evaluation function to predict on the test set
    """
    # Set model to evaluation mode
    # I.e., turn off dropout and set batchnorm to use overall mean and variance (from training), rather than batch level mean and variance
    # Reference: https://github.com/pytorch/pytorch/issues/5406
    model.eval()

    with torch.no_grad():
        tk0 = tqdm(data_loader, total=len(data_loader))
        # Make predictions and calculate loss / jaccard score for each batch
        for bi, d in enumerate(tk0):
            ids = d["ids"]
            token_type_ids = d["token_type_ids"]
            mask = d["mask"]
            orig_ans = d["orig_ans"]
            orig_text = d["orig_text"]
            ans_start = d["ans_start"]
            ans_end = d["ans_end"]
            
            # Move ids, masks, and targets to gpu while setting as torch.long
            ids = ids.to(device, dtype=torch.long)
            token_type_ids = token_type_ids.to(device, dtype=torch.long)
            mask = mask.to(device, dtype=torch.long)
            ans_start = ans_start.to(device, dtype=torch.long)
            ans_end = ans_end.to(device, dtype=torch.long)

            # Predict logits for start and end indexes
            outputs_start, outputs_end = model(
                ids,
                mask,
                token_type_ids
            )
            # Calculate loss for the batch
            loss = loss_fn(outputs_start, outputs_end, ans_start, ans_end)
            # Apply softmax to the predicted logits for the start and end indexes
            # This converts the "logits" to "probability-like" scores
            outputs_start = torch.softmax(outputs_start, dim=1).cpu().detach().numpy()
            outputs_end = torch.softmax(outputs_end, dim=1).cpu().detach().numpy()
            idx_start=np.argmax(outputs_start)
            idx_end=np.argmax(outputs_end)
            out = orig_text[0][int(idx_start)-1:int(idx_end)]
            print(out)
            aaa.append(out)
            with open("res_final.txt","a")as w:
                w.write(out)
                w.write("\n")
```
